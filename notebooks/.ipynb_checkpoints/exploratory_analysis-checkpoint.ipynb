# Members and Lords

First we will import the tables of members of parliament and parties, and merge them.

```python
df_members = psql.read_sql("SELECT * FROM mps", connection)
df_parties = psql.read_sql("SELECT * FROM parties", connection)
```

```python
df_members = df_members.merge(df_parties, left_on='party', right_on='id', how='outer')
```

```python
df_members.head()
```

Let's clean up a few obvious things while we're here.

```python
# get rid of party ids, and urls
df_members.drop(['party_x', 'id_y', 'member_url'], axis=1, inplace=True)
# drop last NaN value
df_members = df_members.iloc[:-2]
# rename party_y to party
df_members.rename(columns={'party_y': 'party'}, inplace=True)
# change end_year 'present' values to current year
df_members['end_year'][df_members['end_year'] == 'present'] = 2017
# convert column to int
df_members['end_year'] = pd.to_numeric(df_members['end_year'], downcast='float')
# change labour co-op to labour
df_members['party'][df_members['party'] == 'Labour (Co-op)'] = 'Labour'
# change member identifier to int
df_members['member_identifier'] = pd.to_numeric(df_members['member_identifier'], downcast='float')
```

We can also calculate a few values we think might be useful.

```python
df_members['term_length'] = df_members['end_year'] - df_members['start_year']
```

Finally we can split the two dataframes into Commons and Lords.

```python
df_lords = df_members[df_members['house'] == 'Lords']
df_commons = df_members[df_members['house'] == 'Commons']
```

# Summary Statistics

We can see that the longest serving Lord in our dataset sat in the house for 78 years, while the longest serving MP sat for 51 years. The mean term lengths of Lords and MPs are 19.45 and 13.51 years respectively, with relatively high standard deviations of 14 and 9.2 years.

```python
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
```

```python
def tokenize(text, stop_words=stop_words, stop=False):
    tokens = [word.strip(string.punctuation) for word in text.split(" ")]
    tokens = [token for token in tokens if len(token) > 0]
    if stop:
        tokens = [t for t in tokens if t not in stop_words]
    tokens = [t.lower() for t in tokens]
    tokens = [t for t in tokens if not t.isdigit()]
    return tokens
```

```python
def lexical_richness(tokens):
    length = len(tokens)
    uniques = len(list(set(tokens)))
    if uniques == 0:
        lr = 0
    else:
        lr = uniques/length
    return lr
```

```python
import gensim
```

```python
def lsi_to_vecs(corpus_lsi):
    lsi_vecs = []

    for c in corpus_lsi:
        vec = [x[1] for x in c]
        lsi_vecs.append(vec)

    return np.array(lsi_vecs)

def tfidf_transform(texts, dictionary=None, tfidf_model=None):
    if not dictionary:
        dictionary = gensim.corpora.Dictionary(texts)
        dictionary.filter_extremes(no_below=5, no_above=0.2)
    corpus = [dictionary.doc2bow(text) for text in texts]
    if not tfidf_model:
        tfidf_model = gensim.models.TfidfModel(corpus)
    corpus_tfidf = tfidf_model[corpus]
    return corpus_tfidf, dictionary, tfidf_model

def lsi_transform(corpus_tfidf, dictionary, lsi_model=None, dimensions=2, topn=25):
    if not lsi_model:
        lsi_model = gensim.models.LsiModel(corpus_tfidf, id2word=dictionary, num_topics=dimensions)
    corpus_lsi = lsi_model[corpus_tfidf]
    lsi_vecs = lsi_to_vecs(corpus_lsi)
    lsi_vecs = lsi_vecs
    return lsi_vecs, corpus_lsi, lsi_modellsi_vecs, corpus_lsi, lsi_model = lsi_transform(corpus_tfidf, 
                                                dictionary)
```
